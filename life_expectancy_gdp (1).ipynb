```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
life = pd.read_csv('all_data.csv')

print(life.isnull().sum())

life.rename(columns={
'Life expectancy at birth (years)': 'Life_expectancy',
}, inplace=True)

print(life.head())

print(life.Country.unique())
```

```python
sns.histplot(data = life, x = life.Year, y = life.Life_expectancy, hue='Country')
plt.show()
plt.clf()
```

```python
sns.scatterplot(data = life, x = 'Year',y='Life_expectancy',hue = 'Country')
plt.show()
plt.clf()
```

```python
sns.lmplot(data=life, x = 'GDP',y='Life_expectancy',hue='Country')
plt.show()
plt.clf()
```

```python
Zimbabwe = life[life.Country =='Zimbabwe']
plt.figure(figsize=(14, 6))

# First subplot for Life Expectancy
plt.subplot(1, 2, 1)
plt.subplot()
sns.regplot(data = Zimbabwe, x = 'Year', y = 'Life_expectancy', color = 'blue',label = 'Life expectancy')
plt.show()
plt.clf()

sns.regplot(data = Zimbabwe, x = 'Year', y = 'GDP', color = 'orange',label = 'GDP')
plt.show()
plt.clf()
```
